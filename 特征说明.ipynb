这段文字包含了几个关于房地产抵押贷款数据字段的定义：

- **dt.first-pi - 第一次支付日期：** 这指的是按照抵押贷款合同条款，第一次定期还款应该支付的日期。日期的格式是YYYYMM，也就是年年年年月月，例如2024年4月表示为202404。

- **flag-fthb - 首次购房者标志：** 这表示贷款人，或者是贷款人群中的一个人，是否为首次购房者。首次购房者是指那些（1）购买抵押财产，（2）将购买的抵押财产作为主要居所，以及（3）在购买抵押财产前的三年内没有对任何住宅产权（不论是独立还是共同）的个人。有一些例外情况，比如被迫搬家的家庭主妇或单亲父母，如果在过去三年里除了与配偶共有的婚内住宅之外没有其他住宅产权的，也可以被考虑为首次购房者。标志的格式是：Y=是，N=否，9=不可用或不适用。

- **dt.matr - 到期日：** 是指抵押贷款笔据上最后一次月付款应支付的月份。日期格式是YYYYMM，例如202404代表2024年4月。

- **cd.msa - 大都市统计区域（MSA）代码：** 用于指明抵押财产所在的大都市统计区域或大都市区的代码，如果为空，则表示该地区既不是一个MSA也不是一个大都市区，或者未知。

- **mi-pct - 抵押贷款保险百分比（MI%）：** 这代表在房利美购买抵押贷款时，抵押贷款保险公司提供的因贷款违约而产生损失的覆盖百分比。如果抵押贷款保险报告的金额少于1%或超过55%，将会显示为“不可用”，即999。如果没有抵押贷款保险，则显示为0。

- **cnt_units - 单元数量：** 表明抵押贷款是一单元、两单元、三单元还是四单元房产，如果为99则表示“不可用”。

- **occupy_sts - 占用状态：** 标明抵押贷款的类型，是自住（P）、第二房（S）还是投资性财产（I），如果为9则表示“不可用”。

- **cltv - 初始组合贷款至价值比（CLTV）：** 如果为999则表示“不可用”。这个比率是指抵押贷款的原始金额与抵押财产价值的组合比率。
这段文字提供了一系列与抵押贷款相关的专业术语的定义：

- **dti - 原始债务收入比（DTI）：** 披露债务收入比，即借款人每月负债总额与总月收入的比率。如果这个比率超过65%，则该数据被标记为“不可用”（999）。

- **orig-upb - 抵押贷款的未偿本金（UPB）：** 在票据日期时，抵押贷款的未偿本金，四舍五入到最近的1,000美元。

- **ltv - 原始贷款至价值比（LTV）：** 取值范围是6% - 105%，如果数据不可用，则标记为999。

- **int_rt - 贷款的利率：** 贷款在发起时票据上注明的利率。

- **channel - 渠道：** 标明贷款是由经纪人、通讯录或第三方发起机构发起的或参与的。如果没有指明经纪人或通讯录，披露将显示“TPO未指定”。类似地，如果既没有第三方发起也没有零售指定，将显示为“零售”。数值代表：R=零售，B=经纪人，C=通讯录，T=TPO未指定，9=不可用。

- **ppmt-pnlty - 提前还款罚金贷款（PPM）标志：** 表示抵押贷款是否是PPM。PPM是指借款人在某些还本事件中被要求支付罚金的抵押贷款。

- **prod_type - 产品类型：** 表明产品是固定利率抵押还是可调利率抵押。

- **st - 状态：** 表示抵押贷款物业所在的州或领土的两字母缩写。

- **prop_type - 物业类型：** 表明抵押贷款抵押物业的类型是公寓（CO），规划单元开发（PU），合作社份额（CP），移动家居（MH）还是单户家庭（SF）。如果物业类型不可用，将标记为99。
这段文字详细描述了与抵押贷款相关的一些术语和数据字段：

- **zipcode - 邮政编码：** 指抵押物业所在地的邮政编码。格式为“#/#/#00”，其中“#/#/#”代表5位邮政编码的前三位数字，而“00”表示未知。

- **id.loan - 贷款唯一标识符：** 分配给每一笔贷款的独特识别号。

- **loan_purpose - 贷款用途：** 表明抵押贷款是现金外债务再融资贷款（C），非指定再融资贷款（R），还是购房贷款（P）。若数据不可用，则标记为9。

- **orig_loan_term - 原始贷款期限：** 抵押贷款的预定月付款次数，从首次付款日期到到期日之间。

- **cnt_borr - 借款人数量：** 必须偿还抵押贷款的借款人数量。仅表明是单一借款人（1）还是多于一个借款人（2）。

- **seller_name - 卖方名称：** 以卖方身份向房利美出售抵押贷款的实体名称。

- **servicer_name - 服务机构名称：** 在最后一个报告周期为房利美提供抵押贷款服务的实体名称。

- **flag_sc - 超级合规标志：** 对于2008年10月1日或之后发起并交付给房利美的超出合规贷款限额的抵押贷款。

- **default - 是否违约：** 标明借款人是否违约的二元回应（即是或否）。
